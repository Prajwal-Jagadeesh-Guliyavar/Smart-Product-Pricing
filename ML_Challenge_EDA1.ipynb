```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import re

print(f'TensorFlow Version: {tf.__version__}')

# --- Configuration ---
DATA_PATH = 'student_resource/dataset/train.csv'
pd.set_option('display.max_colwidth', 80)
sns.set_style('whitegrid')

# --- Load Data ---
print('Loading data...')
df = pd.read_csv(DATA_PATH)
print('Data loaded successfully!')

# --- Initial Inspection ---
print('\n--- Data Info---\n')
df.info()

print('\n--- First 5 Rows---\n')
display(df.head())

print('\n--- Price Statistics---\n')
display(df['price'].describe())

# --- 1. Analyze Price Distribution ---
plt.figure(figsize=(16, 6))

# Histogram of the original price
plt.subplot(1, 2, 1)
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Price', fontsize=15)
plt.xlabel('Price')
plt.ylabel('Frequency')

# Histogram of the log-transformed price
plt.subplot(1, 2, 2)
sns.histplot(np.log1p(df['price']), bins=50, kde=True, color='green')
plt.title('Distribution of Log-Transformed Price', fontsize=15)
plt.xlabel('Log(1 + Price)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# --- 2. Inspect Catalog Content for Feature Extraction ---
print('\n--- Example Catalog Content---\n')
for i in range(5):
    print(f'--- Sample {i+1} ---\n{df["catalog_content"].iloc[i]}\n')

# --- 3. Feature Engineering ---
print('Engineering features...')
# Apply log transform to price
df['log_price'] = np.log1p(df['price'])

def extract_ipq(text):
    # Case-insensitive search for patterns like (Pack of 6), 8 Count, 12 per case., etc.
    patterns = [
        r'pack of (\d+)',
        r'(\d+)\s*per case',
        r'\((\d+)\s*count\)',
        r'pack\s*\((\d+)\)',
        r'(\d+)\s*pack'
    ]

    text_lower = text.lower()
    for pattern in patterns:
        match = re.search(pattern, text_lower)
        if match:
            return int(match.group(1))
    return 1 # Default to 1 if no pack size is found

# Create the IPQ feature
df['ipq'] = df['catalog_content'].apply(extract_ipq)

# --- Verification ---
print('\n--- DataFrame with New Features---\n')
# Displaying relevant columns to check our work
display(df[['price', 'log_price', 'ipq', 'catalog_content']].head(10))

print('\n--- IPQ Value Counts---\n')
display(df['ipq'].value_counts().head(10))

# --- 4. Text Preprocessing ---
def clean_text(text):
    # Remove item name, bullet points, value, and unit headers
    text = re.sub(r'item name:', '', text, flags=re.IGNORECASE)
    text = re.sub(r'bullet point \d+:', '', text, flags=re.IGNORECASE)
    text = re.sub(r'value:', '', text, flags=re.IGNORECASE)
    text = re.sub(r'unit:', '', text, flags=re.IGNORECASE)
    # Convert to lowercase and remove newlines
    text = text.lower().replace('\n', ' ')
    # Remove punctuation and extra whitespace
    text = re.sub(r'[^a-z0-9 ]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['cleaned_text'] = df['catalog_content'].apply(clean_text)

# --- Text Vectorization Setup ---
# The maximum number of words to keep in the vocabulary
MAX_TOKENS = 10000
# The length of sequence to use for each text entry
SEQUENCE_LENGTH = 100

text_vectorizer = tf.keras.layers.TextVectorization(
    max_tokens=MAX_TOKENS,
    output_mode='int',
    output_sequence_length=SEQUENCE_LENGTH
)

# Adapt the vectorizer to our text data to build the vocabulary
print('Adapting text vectorizer...')
text_vectorizer.adapt(df['cleaned_text'])
print('Vectorizer adapted.')

# --- Verification ---
print('\n--- Cleaned Text vs. Original---\n')
display(df[['cleaned_text', 'catalog_content']].head())

print('\n--- Vocabulary Sample---\n')
# First 20 tokens are reserved (0=pad, 1=unknown)
print(text_vectorizer.get_vocabulary()[:20])
```

```python
# --- 5. Image Preprocessing Setup ---
import sys
import os
# Add src directory to path to import utils
if 'student_resource/src' not in sys.path:
    sys.path.append('student_resource/src')
from utils import download_images
from pathlib import Path

# --- Configuration ---
IMAGE_SIZE = (128, 128)
IMAGES_DIR = 'student_resource/images'
os.makedirs(IMAGES_DIR, exist_ok=True)

# --- CORRECTED ---
# The image path should be based on the image URL, not the sample_id
df['image_path'] = df['image_link'].apply(lambda url: os.path.join(IMAGES_DIR, Path(url).name))

def load_and_preprocess_image(path):
    image = tf.io.read_file(path)
    image = tf.io.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, IMAGE_SIZE)
    # Normalize pixels to [0, 1] range
    image = image / 255.0
    return image

# --- Verification ---
# Download and process a SINGLE image to test the pipeline
print('--- Testing image pipeline for one sample ---')
sample_row = df.iloc[0]
sample_id = sample_row['sample_id']
sample_url = sample_row['image_link']
# Get the corrected path
sample_path = sample_row['image_path']

if not os.path.exists(sample_path):
    print(f'Downloading sample image for ID {sample_id}...')
    # The download function in utils.py expects a list of URLs, not (id, url) tuples
    # CORRECTED ARGUMENT:
    download_images([sample_url], IMAGES_DIR)
else:
    print(f'Sample image for ID {sample_id} already exists.')

print(f'Loading and preprocessing image from: {sample_path}')
try:
    processed_image = load_and_preprocess_image(sample_path)
    print('Image processed successfully!')
    print('Shape:', processed_image.shape)
    print('Min value:', tf.reduce_min(processed_image).numpy())
    print('Max value:', tf.reduce_max(processed_image).numpy())

    # Display the image
    plt.imshow(processed_image)
    plt.title(f'Sample ID: {sample_id}')
    plt.axis('off')
    plt.show()
except Exception as e:
    print(f'An error occurred: {e}')
```

```python
import tensorflow as tf

def build_model():
    tf.keras.backend.clear_session()

    # --- Input Layers ---
    image_input = tf.keras.layers.Input(shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3), name='image_input')
    # CORRECTED: Input shape is (1,) and dtype is string for the raw text
    text_input = tf.keras.layers.Input(shape=(1,), dtype=tf.string, name='text_input')
    ipq_input = tf.keras.layers.Input(shape=(1,), name='ipq_input')

    # --- Image Tower ---
    base_image_model = tf.keras.applications.ResNet50(
        include_top=False,
        weights='imagenet',
        input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)
    )
    base_image_model.trainable = False
    image_features = base_image_model(image_input, training=False)
    image_features = tf.keras.layers.GlobalAveragePooling2D(name='image_avg_pool')(image_features)
    image_features = tf.keras.layers.Dense(32, activation='relu', name='image_dense')(image_features)

    # --- Text Tower ---
    # The text_vectorizer layer takes the string input and converts it to numbers
    text_features = text_vectorizer(text_input)
    text_features = tf.keras.layers.Embedding(
        input_dim=MAX_TOKENS,
        output_dim=64,
        mask_zero=True,
        name='text_embedding'
    )(text_features)
    text_features = tf.keras.layers.GlobalAveragePooling1D(name='text_avg_pool')(text_features)
    text_features = tf.keras.layers.Dense(32, activation='relu', name='text_dense')(text_features)

    # --- IPQ Tower ---
    ipq_normalizer = tf.keras.layers.Normalization(axis=None)
    ipq_normalizer.adapt(df['ipq'].values.reshape(-1, 1))
    ipq_features = ipq_normalizer(ipq_input)
    ipq_features = tf.keras.layers.Dense(16, activation='relu', name='ipq_dense')(ipq_features)

    # --- Concatenate Features ---
    concatenated = tf.keras.layers.Concatenate(name='concatenate')([
        image_features,
        text_features,
        ipq_features
    ])

    # --- Regression Head ---
    x = tf.keras.layers.Dense(128, activation='relu', name='combined_dense_1')(concatenated)
    x = tf.keras.layers.Dropout(0.2, name='dropout')(x)
    x = tf.keras.layers.Dense(64, activation='relu', name='combined_dense_2')(x)
    output = tf.keras.layers.Dense(1, name='output')(x)

    # --- Create and Compile Model ---
    model = tf.keras.Model(
        inputs=[image_input, text_input, ipq_input],
        outputs=output
    )
    model.compile(
        optimizer='adam',
        loss='mean_squared_error',
        metrics=[tf.keras.metrics.RootMeanSquaredError(name='rmse')]
    )
    return model

# Re-build the model with the correction
model = build_model()
model.summary()
```

```python
# --- 7. Train the Model on a Sample ---

# Create a small sample of the data for quick training
N_SAMPLE = 1000
df_sample = df.head(N_SAMPLE).copy()

# --- Download images for the sample ---
print(f'Downloading {N_SAMPLE} images for training sample...')
sample_image_urls = df_sample['image_link'].tolist()
# The download function is from utils.py, imported in a previous cell
download_images(sample_image_urls, IMAGES_DIR)
print('Sample images downloaded.')

# --- Create tf.data.Dataset ---
def create_dataset(dataframe):
    # Create the dictionary of inputs
    inputs = {
        "image_input": dataframe['image_path'].values,
        "text_input": dataframe['cleaned_text'].values.reshape(-1, 1), # Reshape for the model input
        "ipq_input": dataframe['ipq'].values.reshape(-1, 1)
    }
    # Get the target variable
    labels = dataframe['log_price'].values

    dataset = tf.data.Dataset.from_tensor_slices((inputs, labels))

    # Define the mapping function to load images
    def map_func(inputs, label):
        inputs['image_input'] = load_and_preprocess_image(inputs['image_input'])
        return inputs, label

    # Apply the mapping function, batch, and prefetch
    return dataset.map(map_func, num_parallel_calls=tf.data.AUTOTUNE).batch(32).prefetch(tf.data.AUTOTUNE)

# Create training and validation splits (80/20)
split_index = int(N_SAMPLE * 0.8)
df_train = df_sample.iloc[:split_index]
df_val = df_sample.iloc[split_index:]

train_dataset = create_dataset(df_train)
val_dataset = create_dataset(df_val)

print(f'\nTraining on {len(df_train)} samples, validating on {len(df_val)} samples.')

# --- Train the model ---
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=10 # Train for 10 epochs on the sample
)

# --- Plot training history ---
plt.figure(figsize=(12, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error (on log_price)')
plt.legend()
plt.show()
```
